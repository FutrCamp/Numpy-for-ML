```
R=np.random.randn(100)
print(R)
print(R.mean())
print(np.mean(R))
print(R.var())
print(R.std())
```

```
R = np.random.randn(1000,3)
print(R)
print(R.mean(axis=0))
print(R.mean(axis=1))
```

np.random.choice(a, size=None, replace= True, p = None)



- Generates random sample from a given 1-D array
- If an ndarray is given, a random sample is generated from its elements.
- If an integer is integer is given, the random sample is generated as if 'a' were np.arange(a)

```
np.random.choice(5,size=(4,4))
```
